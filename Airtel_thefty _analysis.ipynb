```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import re
from datetime import datetime
from textblob import TextBlob
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
from IPython.display import display

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Set visual style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 8)

print("Libraries imported and configured successfully!")
```

```python
# Load the dataset
print("Loading dataset...")
df = pd.read_csv("final_version.csv")

# Check data structure
print(f"Dataset shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
print("\nFirst few rows:")
df.head()
```

```python
# Handle missing values
print(f"Missing values in Text column: {df['Text'].isna().sum()}")
df = df.dropna(subset=['Text'])

# Enhanced text cleaning function
def clean_text(text):
    if pd.isnull(text):
        return ""
    text = re.sub(r"http\S+", "", text)  # remove URLs
    text = re.sub(r"[^a-zA-Z\s0-9]", "", text)  # keep numbers
    return text.lower().strip()

# Apply cleaning
df["clean_text"] = df["Text"].astype(str).apply(clean_text)

print("Text cleaning completed!")
df[['Text', 'clean_text']].head()
```

```python
# Define telecom providers and their keywords
telecom_providers = {
    'mtn': ['mtn', 'mtn mobile money', 'momo', 'mm'],
    'airtel': ['airtel', 'airtel money', 'mm'],
    'vodafone': ['vodafone', 'vodacom', 'mm'],
    'orange': ['orange', 'orange money', 'mm'],
    'tigo': ['tigo', 'tigo cash', 'mm'],
    'safaricom': ['safaricom', 'mpesa', 'mm'],
    'other': ['mobile money', 'mobile wallet', 'ecocash', 'mm']
}

def identify_provider(text):
    for provider, keywords in telecom_providers.items():
        if any(keyword in text for keyword in keywords):
            return provider
    return 'unknown'

# Apply provider identification
df['provider'] = df['clean_text'].apply(identify_provider)

# Filter to only telecom-related tweets
telecom_df = df[df['provider'] != 'unknown'].copy()
print(f"Telecom-related tweets: {len(telecom_df)}")
print(telecom_df['provider'].value_counts())

# Display provider distribution
telecom_df['provider'].value_counts().plot(kind='bar', title='Tweets by Provider')
plt.xticks(rotation=45)
plt.show()
```

```python
# Define complaint categories
complaint_categories = {
    'transaction_issues': ['failed', 'not received', 'not gone through', 'stuck', 'pending', 'reverse'],
    'fraud': ['scam', 'fraud', 'stole', 'theft', 'hack', 'unauthorized'],
    'service_quality': ['slow', 'delay', 'problem', 'error', 'issue', 'not working'],
    'account_issues': ['blocked', 'suspended', 'limit', 'restricted', 'verify'],
    'customer_service': ['customer care', 'support', 'help', 'response', 'agent'],
    'charges': ['charge', 'fee', 'cost', 'deduct', 'expensive'],
    'pin_issues': ['pin', 'password', 'forgot', 'reset', 'remember'],
    'network': ['network', 'connection', 'service', 'signal']
}

def categorize_complaint(text):
    categories = []
    for category, keywords in complaint_categories.items():
        if any(keyword in text for keyword in keywords):
            categories.append(category)
    return categories

# Apply complaint categorization
telecom_df['complaint_categories'] = telecom_df['clean_text'].apply(categorize_complaint)
telecom_df['is_complaint'] = telecom_df['complaint_categories'].apply(lambda x: len(x) > 0)

# Explode categories for analysis
complaint_df = telecom_df.explode('complaint_categories')

print(f"Complaints detected: {telecom_df['is_complaint'].sum()}")
print("Top complaint categories:")
complaint_df['complaint_categories'].value_counts().head(10)
```

```python
def get_sentiment(text):
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity
    if polarity > 0.1:
        return "Positive"
    elif polarity < -0.1:
        return "Negative"
    else:
        return "Neutral"

# Apply sentiment analysis
telecom_df['sentiment'] = telecom_df['clean_text'].apply(get_sentiment)

# Display sentiment distribution
sentiment_counts = telecom_df['sentiment'].value_counts()
plt.pie(sentiment_counts.values, labels=sentiment_counts.index, autopct='%1.1f%%')
plt.title('Sentiment Distribution')
plt.show()
```

```python
def extract_amount(text):
    # Normalize shorthand like 1.4m, 10k, 2bn, etc.
    shorthand_pattern = r'(\d+(?:\.\d+)?)([kKmMbB])'
    shorthand_matches = re.findall(shorthand_pattern, text)
    
    if shorthand_matches:
        num, suffix = shorthand_matches[0]
        num = float(num)
        if suffix.lower() == 'k':
            return int(num * 1_000)
        elif suffix.lower() == 'm':
            return int(num * 1_000_000)
        elif suffix.lower() == 'b':
            return int(num * 1_000_000_000)
    
    # Regular patterns with currencies
    patterns = [
        r'(\d+)\s*(usd|dollars)',
        r'\$(\d+)',
        r'(\d+)\s*(ghs|cedis)',
        r'(\d+)\s*(kes|shillings)',
        r'(\d+)\s*(ugx|shillings)',
        r'(\d+)\s*(frs|francs)',
        r'(\d+)\s*(rands|zar)',
        r'(\d+)\s*(naira|ngn)',
        r'(\d+)\s*(fcfa|xaf)',
        r'lost\s*(\d+)',
        r'stole\s*(\d+)',
        r'(\d+)\s*stolen'
    ]
    
    for pattern in patterns:
        matches = re.findall(pattern, text, flags=re.IGNORECASE)
        if matches:
            amount = matches[0][0] if isinstance(matches[0], tuple) else matches[0]
            try:
                return float(amount)
            except:
                continue
    
    return 0
# Apply amount extraction
telecom_df['amount'] = telecom_df['clean_text'].apply(extract_amount)

# Display financial summary
financial_tweets = telecom_df[telecom_df['amount'] > 0]
print(f"Financial incidents detected: {len(financial_tweets)}")
print(f"Total amount reported: {financial_tweets['amount'].sum():.2f}")
print(f"Average amount per incident: {financial_tweets['amount'].mean():.2f}")

# Display as a scrollable DataFrame with provider included
pd.set_option('display.max_colwidth', None)

display(
    financial_tweets[['provider', 'clean_text', 'amount']].head(800)
    .style
    .set_table_attributes("style='display:inline'")
    .set_table_styles(
        [{'selector': 'table',
          'props': [('max-height', '400px'), ('overflow-y', 'scroll'), ('display', 'block')]}]
    )
)

# Save to CSV including provider
financial_tweets[['provider', 'clean_text', 'amount']].to_csv('financial_tweets.csv', index=False, encoding='utf-8')
print("Saved financial tweets with provider to 'financial_tweets.csv'")
```

```python
# Aggregate total and average lost per provider
provider_summary = financial_tweets.groupby('provider')['amount'].agg(['sum', 'count', 'mean']).reset_index()
provider_summary.rename(columns={'sum': 'total_lost', 'count': 'num_incidents', 'mean': 'average_lost'}, inplace=True)

# Display summary table
display(provider_summary)

# Plot total lost per provider
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(provider_summary['provider'], provider_summary['total_lost'], color='skyblue')
plt.title('Total Lost Amount by Telecom Provider')
plt.xlabel('Provider')
plt.ylabel('Total Lost Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Optional: Plot number of incidents per provider
plt.figure(figsize=(8,5))
plt.bar(provider_summary['provider'], provider_summary['num_incidents'], color='salmon')
plt.title('Number of Financial Incidents by Provider')
plt.xlabel('Provider')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Make sure Date is in datetime format
if 'Date' in complaint_df.columns:
    complaint_df['Date'] = pd.to_datetime(complaint_df['Date'])
    
    # Filter only for August 2025
    complaint_df = complaint_df[
        (complaint_df['Date'].dt.year == 2025) &
        (complaint_df['Date'].dt.month == 8)
    ]
    
    # Get top 5 complaint categories
    top_complaints = complaint_df['complaint_categories'].value_counts().head(5).index.tolist()
    
    # Filter to top complaints only
    top_complaint_df = complaint_df[complaint_df['complaint_categories'].isin(top_complaints)]
    
    # Group by Date (day) and complaint category
    complaints_over_time = top_complaint_df.groupby(
        [top_complaint_df['Date'].dt.date, 'complaint_categories']
    ).size().unstack(fill_value=0)
    
    # Plot
    plt.figure(figsize=(14, 8))
    complaints_over_time.plot(marker='o', ax=plt.gca())
    plt.title('Top Complaint Categories Per Day - August 2025')
    plt.xlabel('Date')
    plt.ylabel('Number of Complaints')
    plt.legend(title='Complaint Type', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig('complaints_over_time.png', dpi=300, bbox_inches='tight')
    plt.show()
else:
    print("No Date column found for temporal analysis.")
```

```python
# Check if we have financial data and date information
if 'Date' in telecom_df.columns and 'amount' in telecom_df.columns:
    # Ensure Date column is datetime
    telecom_df['Date'] = pd.to_datetime(telecom_df['Date'], errors='coerce')
    
    # Filter to tweets with financial amounts
    financial_tweets = telecom_df[telecom_df['amount'] > 0]
    
    if len(financial_tweets) > 0:
        # Filter only August 2025
        august_data = financial_tweets[
            (financial_tweets['Date'].dt.year == 2025) & 
            (financial_tweets['Date'].dt.month == 8)
        ]
        
        if len(august_data) > 0:
            # Group by day
            august_data['day'] = august_data['Date'].dt.day
            daily_impact = august_data.groupby('day')['amount'].agg(['sum', 'count', 'mean']).rename(
                columns={'sum': 'total_amount', 'count': 'incident_count', 'mean': 'average_amount'}
            )
            
            # Create subplots
            fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 12))
            
            # Plot total amount lost by day
            daily_impact['total_amount'].plot(kind='bar', ax=ax1, color='coral')
            ax1.set_title('Total Amount Reported Lost per Day (August 2025)')
            ax1.set_ylabel('Amount (Local Currency)')
            ax1.set_xlabel('Day')
            ax1.tick_params(axis='x', rotation=0)
            
            # Plot average amount lost per incident
            daily_impact['average_amount'].plot(kind='line', ax=ax2, marker='o', color='steelblue')
            ax2.set_title('Average Amount Lost per Incident per Day (August 2025)')
            ax2.set_ylabel('Average Amount (Local Currency)')
            ax2.set_xlabel('Day')
            ax2.tick_params(axis='x', rotation=0)
            ax2.grid(True, alpha=0.3)
            
            plt.tight_layout()
            plt.savefig('financial_impact_august_2025.png', dpi=300, bbox_inches='tight')
            plt.show()
            
            # Print financial summary
            print(f"Total amount reported lost (Aug 2025): {daily_impact['total_amount'].sum():.2f}")
            print(f"Average amount per incident (Aug 2025): {daily_impact['average_amount'].mean():.2f}")
            print(f"Total financial incidents (Aug 2025): {daily_impact['incident_count'].sum()}")
        else:
            print("No financial incidents found for August 2025.")
    else:
        print("No financial incidents found in the dataset.")
else:
    print("Date or amount information not available for financial analysis.")
```

```python
# Extract user-specific issues
user_issues = {
    'pin_issues': ['forgot pin', 'pin reset', 'cannot remember pin', 'change pin', 'pin problem'],
    'registration': ['cannot register', 'verify account', 'registration problem', 'account activation'],
    'app_issues': ['app not working', 'app crash', 'app problem', 'update app'],
    'cashout_issues': ['cannot withdraw', 'cashout problem', 'withdrawal issue', 'agent has no money'],
    'deposit_issues': ['cannot deposit', 'deposit problem', 'money not reflecting'],
    'transaction_limit': ['limit exceeded', 'transaction limit', 'increase limit']
}

def detect_user_issues(text):
    issues = []
    for issue, keywords in user_issues.items():
        if any(keyword in text for keyword in keywords):
            issues.append(issue)
    return issues

# Apply user issue detection
telecom_df['user_issues'] = telecom_df['clean_text'].apply(detect_user_issues)
user_issues_df = telecom_df.explode('user_issues')

# Plot user-centered issues
plt.figure(figsize=(12, 8))
user_issue_counts = user_issues_df['user_issues'].value_counts()
user_issue_counts.plot(kind='barh', color='lightseagreen')
plt.title('User-Centered Issues Reported')
plt.xlabel('Number of Reports')
plt.tight_layout()
plt.savefig('user_issues.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
def extract_phrases(text, n=2):
    tokens = word_tokenize(text)
    phrases = []
    for i in range(len(tokens) - n + 1):
        phrases.append(" ".join(tokens[i:i+n]))
    return phrases

# Extract bigrams from complaints
all_phrases = []
for text in complaint_df['clean_text']:
    all_phrases.extend(extract_phrases(text, 2))

# Count and display most common phrases
phrase_counts = Counter(all_phrases).most_common(20)

print("Most common phrases in complaints:")
for phrase, count in phrase_counts:
    print(f"{phrase}: {count}")

# Visualize top phrases
phrases_df = pd.DataFrame(phrase_counts, columns=['Phrase', 'Count'])
plt.figure(figsize=(12, 8))
plt.barh(phrases_df['Phrase'], phrases_df['Count'])
plt.title('Top 20 Phrases in Complaints')
plt.xlabel('Frequency')
plt.tight_layout()
plt.show()
```
